* Genre 

```python
fig,ax=plt.subplots(figsize=(15,8))
sns.countplot(data=df,x="genre",order=df["genre"].value_counts().index)
plt.xlabel("Genre")
plt.ylabel("No.of movies")
plt.title("Movies genre frequency")
```

```python
avg_score_for_each_genre=pd.DataFrame(df.groupby("genre")["score"].mean().sort_values(ascending=False))
avg_score_for_each_genre
```

* Movies Rating

```python
fig,ax=plt.subplots(figsize=(8,6))
sns.countplot(data=df,x="rating",order=df["rating"].value_counts().index)
plt.xlabel("rating")
plt.ylabel("No.of movies")
plt.title("Movies rating frequency")
```

```python
avg_score_for_each_rating=pd.DataFrame(df.groupby("rating")["score"].mean().sort_values(ascending=False))
avg_score_for_each_rating
```

* Movies popularity based on votes

```python
top_10_popular=df.sort_values("votes",ascending=False).head(10).reset_index()
top_10_popular
```

* directors list arranged based on number of movies and avg score

```python
dir=pd.DataFrame(df.groupby("director").agg({"score":["count","mean"]}))
dir.columns=["No.Movies","Avg.Score"]
dir.sort_values("No.Movies",ascending=False)
```

```python
dir.sort_values("Avg.Score",ascending=False)
```

* stars list arranged based on number of movies and avg score of their movies

```python
star_list=pd.DataFrame(df.groupby("star").agg({"score":["count","mean"]}))
star_list.columns=["No.Movies","Avg.Score"]
star_list.sort_values("No.Movies",ascending=False)
```

```python
star_list.sort_values("Avg.Score",ascending=False)
```

* countries list arranged based on number of movies and avg score of their movies

```python
countries_list=pd.DataFrame(df.groupby("country").agg({"score":["count","mean"]}))
countries_list.columns=["No.Movies","Avg.Score"]
countries_list.sort_values("No.Movies",ascending=False).head(10)
```

```python
countries_list.sort_values("Avg.Score",ascending=False).head(10)
```

* companies production shares,profit and expendetures

```python
companies_list=pd.DataFrame(df.groupby("company").agg({"score":["count","mean"],"gross":"sum","budget":"sum"}))
companies_list.columns=["No.Movies","Avg.Score","total revemue","total expenses"]
companies_list["net profit"]=companies_list["total revemue"]-companies_list["total expenses"]
companies_list.sort_values("No.Movies",ascending=False)
```

#### Correlation

* Years Vs Score (Do movies get better or worse?)

```python
annual_score=pd.DataFrame(df.groupby("year")["score"].mean())
fig,ax=plt.subplots(figsize=(8,6))
sns.scatterplot(x=annual_score.index,y=annual_score["score"])
plt.xlabel("Year")
plt.ylabel("Avg. Score")
plt.title("Movies score over years")
```

* Votes VS. Score (Does low number of votes mean that the movie sucks?)

```python
fig,ax=plt.subplots(figsize=(8,6))
sns.scatterplot(x=df["votes"],y=df["score"])
plt.xlabel("Number of votes in millions")
plt.ylabel("Score")
plt.title("Number of votes VS. movie score")
```

* Budget VS. Score (Can a low budget movie get a pretty good score? and Is the high budget of the movie the key to success?)

```python
fig,ax=plt.subplots(figsize=(8,6))
sns.scatterplot(x=df["budget"],y=df["score"])
plt.xlabel("budget in 100 millions")
plt.ylabel("Score")
plt.title("movie budget VS. movie score")
```

* Revenue and budget (Is the movie industry a good or risky investment?)

```python
fig,ax=plt.subplots(figsize=(8,6))
sns.scatterplot(x=df["budget"],y=df["gross"])
plt.xlabel("budget in 100 millions")
plt.ylabel("gross revenue")
plt.title("movie budget VS. gross revenue")
```

* Revenue and movie score (Is a good movie a profitable movie? Are there really a commercial movies?)

```python
fig,ax=plt.subplots(figsize=(8,6))
sns.scatterplot(x=df["gross"],y=df["score"])
plt.xlabel("gross revenue")
plt.ylabel("movie score")
plt.title("gross revenue VS. movie score")
```

* Overall Correlation

```python
corr=df.corr()
corr
```

```python
fig,ax=plt.subplots(figsize=(8,6))
heat_map=sns.heatmap(corr,annot=True,cmap='magma_r',linewidths=0.2)
plt.title("correlations heat map")
```

```python
corr_pair=corr.unstack()
corr_table=pd.DataFrame(corr_pair.sort_values(ascending=False))
corr_table.columns=["correlation"]
corr_table[corr_table["correlation"]!= 1]
```
